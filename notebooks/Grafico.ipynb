```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
# Opciones de pandas para ver más info
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)


# Ruta a tus datos (ajústala según tu carpeta destino)
DATA_PATH = r"C:\Users\migue\OneDrive\Escritorio\Fraude_bancario\data\creditcard.csv"

# 1. Cargar
df = pd.read_csv(DATA_PATH)

# 2. Resumen
print(df.head(5))
```

```python
# Forma y tipos
print("Dimensiones:", df.shape)
print("\nTipos de datos:\n", df.dtypes.value_counts())

# Resumen estadístico
df.describe().T.head(10)

# Valores nulos
df.isna().sum().sort_values(ascending=False).head(10)

# Duplicados
print("Duplicados:", df.duplicated().sum())
```

```python
df.describe()
```

```python
# Balance de clases
sns.countplot(x="Class", data=df)
plt.title("Distribución de Fraude (1) vs No Fraude (0)")
plt.show()

print(df["Class"].value_counts(normalize=True))

# Tasa de fraude
if 1 in df["Class"].unique():
    fraud_rate = (df["Class"] == 1).mean()
    print(f"Tasa de fraude: {fraud_rate:.4%}")
```

```python
sns.histplot(df["Amount"], bins=100, kde=True)
plt.title("Distribución de Montos de Transacciones")
plt.show()
```

```python
plt.figure(figsize=(12,8))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, cmap="coolwarm", center=0)
plt.title("Mapa de Correlaciones")
plt.show()
```

```python
# Ejemplo: Variable 'V1' por clase
def graf_bigote(Variable , Etiqueta = "Class" , df = df ):
    sns.boxplot(x = Etiqueta , y = Variable , data = df , showfliers = False)
    plt.title(f"Distribución de {Variable} según clase")
    plt.show()
```

```python
for i in df.columns:
    graf_bigote(i)
    
```
